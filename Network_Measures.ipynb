**Plot the degree distribution of the 3 networks below.**

```python
G1=nx.erdos_renyi_graph(10000,40/10000)
G2=nx.watts_strogatz_graph(10000,40,0.05)
G3=nx.barabasi_albert_graph(10000,20)
```

```python
import collections
# List of graphs
graphs = [G1, G2, G3]
graph_names = ["Erdos-Renyi", "Watts-Strogatz", "Barabasi-Albert"]

# For each graph
for i, G in enumerate(graphs):
    plt.figure(i)
    degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
    degreeCount = collections.Counter(degree_sequence)
    deg, cnt = zip(*degreeCount.items())

    plt.bar(deg, cnt, width=0.80, color='b')

    plt.title(f"Degree Histogram of {graph_names[i]}")
    plt.ylabel("Count")
    plt.xlabel("Degree")
    plt.show()
```

# 2. Clustering Coefficient

```python
G=nx.Graph()
G.add_edges_from([(0,1),(1,2),(2,3),(0,2)])
nx.draw(G,with_labels=True)
```

```python
list(G.neighbors(2))
```

node 2 has 3 neighbors. (1,3,0)


```python
print(G.has_edge(1,3))
print(G.has_edge(1,0))
print(G.has_edge(3,0))
```

node 1 and 0 are neighbor each other.

So node 0, 1, and 2 forms a cluster (geometrically, this is triangle)

Maximum number of cluster at node 2 is 3, but there is only 1 cluster.

So clustering coefficient at node 2 is 1/3

```python
nx.clustering(G)
```

```python
nx.average_clustering(G)
```

**Calculate average clustering coefficient of the 3 networks(G1, G2, G3).**

```python
# Calculate average clustering coefficient
avg_clustering_G1 = nx.average_clustering(G1)
avg_clustering_G2 = nx.average_clustering(G2)
avg_clustering_G3 = nx.average_clustering(G3)

# Print the results
print(f"Average clustering coefficient of G1: {avg_clustering_G1}")
print(f"Average clustering coefficient of G2: {avg_clustering_G2}")
print(f"Average clustering coefficient of G3: {avg_clustering_G3}")
```

**Plot the histogram of clustering coefficient of the 3 networks(G1, G2, G3).**

```python
# For each graph
for i, G in enumerate(graphs):
    plt.figure(i)
    clustering_coeffs = nx.clustering(G).values()
    plt.hist(clustering_coeffs, bins=20)
    plt.title(f"Clustering Coefficient Histogram of {graph_names[i]}")
    plt.ylabel("Count")
    plt.xlabel("Clustering Coefficient")
    plt.show()
```

**Calculate the average shortest path length of the 3 networks(G1, G2, G3).**

hint: Randomly choose n number of pair of nodes, and then calculate average shortest path length. If you calculate path length for every pair of nodes, it will take lots of time.

```python
import random

def average_shortest_path_length(G, num_samples=1000): #more pairs of nodes will generally give more accuracy on the estimation of average shortest path length
    nodes = list(G.nodes)
    path_lengths = []
    for _ in range(num_samples):
        u, v = random.sample(nodes, 2)
        try:
            path_lengths.append(nx.shortest_path_length(G, u, v))
        except nx.NetworkXNoPath:
            continue
    return sum(path_lengths) / len(path_lengths)

# Calculate average shortest path length
avg_path_length_G1 = average_shortest_path_length(G1)
avg_path_length_G2 = average_shortest_path_length(G2)
avg_path_length_G3 = average_shortest_path_length(G3)

# Print the results
print(f"Average shortest path length of G1: {avg_path_length_G1}")
print(f"Average shortest path length of G2: {avg_path_length_G2}")
print(f"Average shortest path length of G3: {avg_path_length_G3}")
```

**Let's say there are about 30000 students and faculties in a sample university. Suppose we have 30 number of friends or acquaintances on average. If friendship network follows watts_strogatz_graph with p=0.05, how many degrees of separation we need?**

```python
# Generate the Watts-Strogatz graph
G = nx.watts_strogatz_graph(30000, 30, 0.05)

# Calculate the average shortest path length (average degrees of separation)
avg_degrees_of_separation = average_shortest_path_length(G, num_samples=1000)

print(f"Average degrees of separation: {avg_degrees_of_separation}")
```
